```python
import pandas as pd
```

```python
# Load CSV as it is
root = '/content/drive/MyDrive/Program/Python/Data ML DL/Datasets/data_cleaning_challenge.csv'
read_csv = pd.read_csv(root)
```

```python
# Drop an empty columns
drop_empty_col = read_csv.drop(columns=['Unnamed: 9', 'Unnamed: 10'])
```

```python
# Drop NaN value
drop_nan_row = drop_empty_col[drop_empty_col['Row Type'].notna()]
```

```python
# Number of data iterations
col_values = []
iteration = 0
for i in drop_nan_row['Row Type']:
    if 'first name' in i:
        iteration += 1
    col_values.append(iteration)
```

```python
# Add new column and its value
add_iteration_col = drop_nan_row
add_iteration_col['Iterations'] = col_values
```

```python
# Exclude row with 'Row Type' as its value
drop_extra_row = add_iteration_col[add_iteration_col['Row Type'] != 'Row Type']
```

```python
# Make a dataframe consist of name only
name_dataframe = drop_extra_row[drop_extra_row['Row Type'].str.contains('first name')]
```

```python
# Slice it to make it prettier
name_dataframe = name_dataframe.loc[:, ['Row Type', 'Iter Number', 'Power1', 'Iterations']] # Filter the columns

name_dataframe.rename(columns={'Row Type': 'First Name', 'Iter Number': 'Last Name', 'Power1': 'Date'}, inplace=True)
name_dataframe['First Name'] = name_dataframe['First Name'].str[12:]
name_dataframe['Last Name'] = name_dataframe['Last Name'].str[10:]
name_dataframe['Date'] = name_dataframe['Date'].str[5:]
name_dataframe
```

```python
# Exclude any rows that contain 'first name' as part or entirity of its value with '~' argument
no_name_dataframe = drop_extra_row[~drop_extra_row['Row Type'].str.contains('first name')]
no_name_dataframe
```

```python
# Merge it, order by 'Iterations'
pd.merge(left=name_dataframe, right=no_name_dataframe, how='inner', on='Iterations')
```
