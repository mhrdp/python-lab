# Standart Library Collections

To import this libary wrote: <code>import collections</code>, or you can also do: <code>from collections import <i>function_name</i></code>.

<ol>
  <li><code>collections.Counter</code> <a href="#scrollTo=ydTXLvKoRlwt&line=1&uniqifier=1">go to cell</a><br>Is a container in the <code>collections</code> class to store the items as a dictionary key and the count of those items as a dictionary value.<br>Return value: <code>dict</code></li>
  <li><code>collections.defaultdict</code> <a href="#scrollTo=wOjoPmzGZO5W&line=6&uniqifier=1">go to cell</a><br>Is a container in the <code>collection</code> class that pretty much the same as dictionary (dict) container. The only different is that this container will combine dictionary with the same key as one, make it have a multiple value<br>Return value: <code>dict</code> with maybe a multiple values assigned for one key</li>
  <li><code>collections.namedtuple</code> <a href="#scrollTo=X8pSJtxna9ak&line=2&uniqifier=1">go to cell</a><br>Is an easy to create, lightweight object type that basically turn tuple into a convinient container for a simple tasks. With <code>namedtuple</code> you don't need to use integer indices to access tuple's items<br>Return value: <code>tuple</code></li>
  <li><code>collections.OrderedDict</code> <a href="#scrollTo=N2DpbbTaHRvT&line=1&uniqifier=1">go to cell</a><br>Is a way to create a dictionary that will remember the order it was created when you iterate through it. In a normal dictonary, the iterations are in a arbitrary/random order.<br>Return value: <code>dict</code></li>
</ol>

```python
from collections import Counter

my_list = [1, 1, 1, 2, 3, 3]
my_list_counter = Counter(my_list)

print(f'collection.Counter result: {dict(my_list_counter)}.')
```

```python
from collections import defaultdict

init_default = defaultdict(list)
init_default['python'].append('programming languange')
init_default['C'].append('systemic languange')
init_default['python'].append('simplistic')
init_default['PHP'].append('the persistent old folk')

init_default
```

```python
from collections import namedtuple

# Example 1
Point = namedtuple('Point', 'x, y')

# Point(x, y)
pt1 = Point(1, 2)
pt2 = Point(3, 4)
sum_val = (pt1.x * pt2.x) + (pt1.y * pt2.y)
print(pt1)
print(pt2)
print('(1 * 3) + (2 * 4) =', sum_val)

print('\n')


# Example 2, notice the absence of commas ( , ) when defining the name of tuple's value
Car = namedtuple('Car', 'Name Mileage Colour Class')

my_car = Car(Name='Toyota', Colour='Silver', Mileage=200000, Class='Yaris')
my_car = Car(Name='Honda', Class='Jazz', Mileage=50000, Colour='Red') # This will rewrite previous value
print('Value without asterisk ( * ):', my_car)
print('Value with asterisk ( * ):', *my_car)
print('My car name is:', my_car.Name)
```

```python
from collections import OrderedDict

# in an OrderedDict, the iteration will be executed according to the order of the items when it was inserted
ordered_dictionary = OrderedDict()
ordered_dictionary['a'] = 1
ordered_dictionary['c'] = 3
ordered_dictionary['b'] = 2

print(ordered_dictionary)

ordinary_dictionary = {}
ordinary_dictionary['a'] = 1
ordinary_dictionary['c'] = 3
ordinary_dictionary['b'] = 2

print('Ordinary dictionary:', ordinary_dictionary)
```
