```python
# 'Unpack' dictionary into separated value using dict.items()

my_dict = {'C': 'Programming languange',
           'Microsoft': 'Operating system written in C'}
new_dict = {}
for key, value in my_dict.items():
  print('my_dict\'s key:', key)
  print('my_dict\'s value:', value)
  new_dict[key] = value
print('The new appended dict:', new_dict)
```

```python
# Make a tuple based on specified element's last occurence inside the string using string.rpartition()
s = 'Hey you, Moona'

#separate the element based on the last occurence of ' '
separate = s.partition(' ')
print('Result after separation: ', separate)

print('\n')

string_1, space, name = s.rpartition(' ')
print('Result after you clean the data showed below', 'first element:' + string_1, 'element that was separated (a space):' + space, 'last element:' + name, sep='\n')
```
