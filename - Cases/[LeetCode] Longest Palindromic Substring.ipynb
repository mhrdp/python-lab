```python
s1 = 'babad'
s2 = 'cbbd'
s3 = 'a'
s4 = 'ac'

def longestPalindrome(s):
    # init res
    res = ''

    for i in range(len(s)):

        # odd -> helper, update
        tmp = helper(i, i, s)
        if len(res) < len(tmp):
            res = tmp

        # even -> helper, update
        tmp = helper(i, i + 1, s)
        if len(res) < len(tmp):
            res = tmp

    # return res
    return res

def helper(l, r, s):

    # if inbound and palindrome, move left left and right right
    while (l >= 0 and r < len(s) and s[l] == s[r]):
        l -= 1
        r += 1

    # return 
    return s[l + 1: r]

print(longestPalindrome(s1))
print(longestPalindrome(s2))
print(longestPalindrome(s3))
print(longestPalindrome(s4))
```
