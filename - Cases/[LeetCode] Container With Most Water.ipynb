**Problems**:<br>
Given `n` non-negative integers `a1, a2, ..., an` , where each represents a point at coordinate `(i, ai)`. n vertical lines are drawn such that the two endpoints of the line `i` is at `(i, ai)` and `(i, 0)`. Find two lines, which, together with the x-axis forms a container, such that the container contains the most water.

**Notice** that you may not slant the container.

```python
def max_area(height):
    if not height:
        return 0
    
    length = len(height)
    start = 0
    end = length - 1

    begin = (end - start) * min(height[start], height[end])

    while start < end:
        if height[start] < height[end]:
            current_low = height[start]
            start += 1
        
            while height[start] < current_low:
                start += 1

            if (end - start) * min(height[start], height[end]) > begin:
                begin = (end - start) * min(height[start], height[end])

        else:
            current_high = height[end]
            end -= 1
            
            while height[end] < current_high:
                end -= 1
            
            if (end - start) * min(height[start], height[end]) > begin:
                begin = (end - start) * min(height[start], height[end])
        return begin
```

```python
height1 = [1,8,6,2,5,4,8,3,7]
height2 = [4,3,2,1,4]
height3 = [1,2,1]
height4 = [1,1]

print(max_area(height1))
print(max_area(height2))
print(max_area(height3))
print(max_area(height4))
```
