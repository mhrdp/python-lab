**Problem**:<br>
You are given an **integer** array `list_rank` of size `n`, where `list_rank[i]` is the score of the i-th athlete in a competition. All the scores are **guaranteed to be unique**.

The athletes are placed based on their scores, where the 1st place athlete has the highest score, the 2nd place athlete has the 2nd highest score, and so on. The placement of each athlete determines their rank:

+ The 1st place athlete's rank is **"Gold Medal"**.
+ The 2nd place athlete's rank is **"Silver Medal"**.
+ The 3rd place athlete's rank is **"Bronze Medal"**.
+ For the 4th place to the n-th place athlete, their rank is their placement number (i.e., the xth place athlete's rank is "x").

Return an array answer of size `n` where `answer[i]` is the rank of the i-th athlete.

# Explanation
The key point here is to use `dict` and `sorted` dictionary.
1. First we will initialize empty dictionary: `rank_dict = {}`; and set current ranking to `0`: `rank = 0`
2. Then we will iterate through the `list` inside the function argument with `for` loop, and append it to `rank_dict`
3. After that we will sort `rank_dict` from big to small with `sorted()[::-1]`. We will assign it to `sorted_rank_dict` variable.
    - **Note**: After `dict` object being `sorted()`, it will become `list` object with `dict's key` is its values. See [here](#scrollTo=8qtKskg5UBno&line=2&uniqifier=1)
4. Then we will use `try.. except..` module to begin determine the medal of each player. Inside `try`, we will:
    - Use `sorted_rank_dict` first value to third value (index no. 0 to no. 2) as a `dict's key` for `rank_dict` dictionary, and change it values from `0` to `'Gold Medal'`, `'Silver Medal'`, `'Bronze Medal'` respectively.
    - And don't forget to specify the rank placement for each respective medal.
5. In `except` module, we will just `pass`
6. Increment the `rank` by 1: `rank += 1`. This is needed because the next rank is `4`, while before this point the `rank` value is still `3`.
7. Lastly, we will iterate through `sorted_rank_dict` with `for` loop with `j` as its variable. Inside it we will:
    - Add `if` condition that will take `rank_dict[j] == 0` as its argument. Meaning that, if `rank_dict[j]` value is equal to `0`, which means every value **above rank 3**.
    - If true, we will change its value by current `rank`.
8. Return `rank_dict.values()`.


```python
def find_relative_rank(list_rank):
    rank_dict = {}
    rank = 0

    for i in list_rank:
        rank_dict[i] = 0
    
    sorted_rank_dict = sorted(rank_dict)[::-1]

    try:
        rank_dict[sorted_rank_dict[0]] = 'Gold Medal'
        rank = 1

        rank_dict[sorted_rank_dict[1]] = 'Silver Medal'
        rank = 2

        rank_dict[sorted_rank_dict[2]] = 'Bronze Medal'
        rank = 3
    except:
        pass
    
    rank += 1

    for j in sorted_rank_dict:
        if rank_dict[j] == 0:
            rank_dict[j] = str(rank)
            rank += 1
    return rank_dict.values()

list_a = [5,4,3,2,1]
list_b = [10,3,8,9,4]

print('List a:', find_relative_rank(list_a))
print('List b:', find_relative_rank(list_b))
```

## About `sorted()` when combined with `dict`
It seems like when you sorted `dict` object using `sorted()`, it will convert into `list` object with `dict's keys` as its value. 

```python
list_a = [5,6,3,2,1]

d = {}
for i in list_a:
    d[i] = 0

print('Dictionary object before .sorted():', d)
print('Dictionary object after .sorted():', sorted(d)[::-1])
```
