# Explanations



1. First we will initialize the <code>prefix</code> list, to later append the common prefix to show the result
2. Then we will iterate through all the items inside the list, few things to note here: <br>
  - <code>\*</code> symbol used in <code>zip(\*<i>list_of_string</i>)</code> is to 'unzip' the list before being 'zipped'
  - When the 'unzipped' list being 'zipped' again with <code>zip()</code>, it will return the first letter in the collection of words -- that previously a list -- as a row, make it vertically sorted instead of horizontally. Look at the example here: <a href="N#scrollTo=ZFPp46ag2OAO&line=4&uniqifier=1">example</a>
3. On each iteration, using <code>if</code> conditional, find a first few iterations that return exactly one (1) letter and immediatedly <code>break</code> the loop once it return more that one (1) letter. To make it clearer, here's the logic:
  - We will be using <code>set()</code> method to combine all the the similar letter in each iteration, and count the length of the returned set with <code>len()</code> method.
  - If the <code>len()</code> is anything more than one (1), that means that iteration was not compromised of a common letters between each word, which is not what we want so it will trigger <code>break</code> parameter to stop the program.
  - If the <code>len()</code> is exactly one (1) that mean we find a common letter, and it will continue to the next iteration until the <code>len()</code> did not return one (1).
4. For each iteration, we will append the <code>True</code> value to a new list and convert it into a string using <code>''.join(<i>iterable</i>)</code> method.





```python
list_of_string = ['flow', 'flower', 'flick']

prefix = []
for i in zip(*list_of_string):
  if len(set(i)) != 1:
    break
  else:
    prefix.append(i[0])
print(''.join(prefix))
```

```python
list_of_string = ['flow', 'flower', 'flick']

print('Unzipped result with * :')
print(*list_of_string)

print('\nResult after unzipped and zipped back:')
for i in zip(*list_of_string):
  print(i)
```
