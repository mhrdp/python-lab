**Problem**:<br>
The string `PAYPALISHIRING` is written in a zigzag pattern on a given number of rows like this: (you may want to display this pattern in a fixed font for better legibility)
```
P   A   H   N
A P L S I I G
Y   I   R
```
And then read line by line: `PAHNAPLSIIGYIR`

Write the code that will take a string and make this conversion given a number of rows:
```
string convert(string s, int numRows);
```

**Solution**:<br>
- Runtime: 48ms - Faster than 96.16% of submissions
- Memory: 14.3mb - More efficient than 88.43% of submissions

**Complexity**:
- Time complexity: O(n)
- Space complexity: O(n)

### For explanation look below the code

**Important notice**: Take a look at the pattern of the end result of the example in the case above. It move in a zig zag manner from top down, then diagonally to the top right, then go down again to repeat the pattern.

````python
def convert(s, numRows):
    if numRows == 1 or numRows >= len(s):
        return s
        
    L = [''] * numRows # Create number of index based on numRows argument
    idx, step = 0, 1
    
    for x in s:
        L[idx] += x # increment L[idx] with x ---> L[idx] = L[idx] + x
        if idx == 0:
            step = 1
        elif idx == numRows - 1:
            step = -1
        idx += step
            
    return ''.join(L)
````

````python
str1 = 'PAYPALISHIRING'

print(convert(str1, 3))
````

# Explanation
Some explanation about the code.
```
if numRows == 1 or numRows >= len(s):
        return s
```
This code is to return the argument `s` if the input not valid

```
L = [''] * numRows
```
This code basically telling the `list` `L` to repeat `''` (empty value) for `numRows` amount of time.
___

```
idx, step = 0, 1
```
`idx` basically means *index*, implying the number of list's index. `idx` has to be `<= numRows-1`. While `step` will serve as the index for which `s` argument will be placed, the rule of `idx` will also applied to this.

### Now let's break down the `for` loop
The function of this `for` loop is to basically iterate through the `s` argument and fill it into `L` list. We will control how to fill the list by adding `if` conditionals. Essentially, what we want is for the list to be filled from first index -> last index -> first_index -> ... and so on and so forth until the iteration over.

```
for x in s:
    L[idx] += x
    if idx == 0:
        step = 1
    elif idx == numRows - 1:
        step = -1
    idx += step
```
There are several component here inside. All in all the general concept pretty simple but can be quite confusing. But we will walk through it one-by-one.

```
for x in s:
```
We will declare `for` loop to iterate through the `s` argument, which is a `str`. The iteration will be saved to `x`

```
L[idx] += x
```
This will increment L (which is an empty list times `numRows`) with `x` value. Increment mean it won't overrides the old value, but instead append it for each iteration.

```
if idx == 0:
    step = 1
```
This code basically telling, 'if `idx` equal to `0`, change `step` value to `1`. And remember the `idx` rule explained before, that `idx` can't be higher than `numRows`

```
if idx == numRows-1:
    step = -1
```
This code function the same as above, the difference is in its condition and the value of `step`. If the `idx` already reached it maximum value, substract `step` by `-1`.

```
idx += step
```
Now this is the finishing touch. This code basically telling `idx` to increment itself by `step`. So when `step = 1`, which mean it start when `idx` is `0`, it will keep increase `idx` value by `1`; and when `idx` reach `numRows-1` (last index) the `step` will become `-1` and it will substract `idx` value by `-1` until `idx` reach `0` again and `step` became `1` once again. It will repeat those cycles until the iteration is over.

When `idx` became `-1` it will basically fill the `L` list in a backward manner, meaning it will fill the list from the last index to the first index, and vice versa.
