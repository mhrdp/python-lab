# Two Sum II

## Problem Statement
Given an array of integers `numbers` that is already sorted in non-decreasing order, find two `numbers` such that they add up to a specific target number.

Return the indices of the two `numbers` (1-indexed) as an integer array answer of size 2, where `1 <= answer[0] < answer[1] <= numbers.length`.

The tests are generated such that there is exactly one solution. You may not use the same element twice.

# twoSum1 Explanation
The way this function work is by adding the `numbers` value as a `key` in the `Hash` dictionary with its index as its dict's value if it did not met the requirement of:
```
if target - numbers[i] in Hash.keys()
```
If the requirement was met, the dictionary will fetch its value by taking `Hash[target-numbers[i]]` as its key. The `Hash[target-numbers[1]]` will return the key's **real** index value, so we need to add `+ 1` because we will use 1-index, not 0-index. We also need to add `i + 1` as a second items, with `i` being the current `index` in the `for` iterations. Anyway, the `return` will become like this:
```
return [Hash[target-numbers[i]] + 1, i + 1]
```

````python
def twoSum1(numbers, target):
    Hash = {}
    for i in range(0, len(numbers)):
        if target - numbers[i] in Hash.keys():
            return[Hash[target-numbers[i]] + 1, i + 1]
        else:
            Hash[numbers[i]] = i
    return []
````

# twoSum2 Explanation
This pretty straight forward, it basically use **two pointer method**, with `left` as the left-most value, and `right` as the right-most value. We will shrink the scope if `numbers[left] + numbers[right] > target` or `numbers[left] + numbers[right] < target` until it become equal to `target`.

````python
def twoSum2(numbers, target):
    left = 0
    right = len(numbers)-1
    while left < right:
        if numbers[left] + numbers[right] == target:
            return [left+1, right+1]
        elif numbers[left] + numbers[right] > target:
            right -= 1
        else:
            left += 1
````
