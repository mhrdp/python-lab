# Explanation
We will utilize simple `for` and `while` loop in this case.

## Step by step
1. Check whether the length of `s` == 0, if so return 0
2. Initialize `answer` and `j` to `1`
3. `for i in range(len(s))` - iterate through the whole string `s`, inside it:
    - Initialize `while` loop taking `len(s)` as its upper limit with `j` as its base value
    - Also, add `and` conditional to make sure `s[j]` is not in `s[i:j]`
    - Lastly, add `j += 1` so it won't looping endlessly
    - All of the above is to make sure the loop **does not** stop as long as `j` still under `len(s)` **and** `s[j]` value **is not** within `s[i:j]`
4. Overwrite `answer` with the `max` value between (old) `answer` and `j - i`
    - `j - i` is to get the range between the highest index and the lowest index
5. Return `answer`
