```python
pascal_depth = 5
def find_pascal(depth):
    default_list = [1]
    result = []
    for _ in range(depth):
        result.append(default_list)
        default_list = count_pascal(default_list)
    return result

def count_pascal(row):
    new_list = []
    new_list.append(1)
    for i in range(len(row)-1):
        counting = row[i] + row[i+1]
        new_list.append(counting)
    new_list.append(1)
    return new_list
print(find_pascal(pascal_depth))
```
