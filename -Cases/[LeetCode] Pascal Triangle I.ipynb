# Generate Pascal Triangle
In this case, we will use **two functions** to solve this problem, for readibility. The main function will take the number of depth of the Pascal tree, and the second function will be responsible to count the pascal value for each row.

1. `find_pascal()`, the main function:
  - `default_list` : add one (1) as its value as a placeholder value for all of the Pascal value later on
  - `result` : a empty list to store the result later on
  - `for _ in range(depth)` : iterate *'depth'* number of times, we use `_` because we won't use its value
  - `result.append(default_list)` : we add the `default_list` value into `result` list
  - `default_list = count_pascal(default_list)` : update the `default_list` variable by calling `count_pascal()` function for the next iteration
  - `return result` : return the `result` list

2. `count_pascal()`, the function to count the pascal number for each row:
  - `new_list` : initialize a new list to insert the pascal values
  - `new_list.append(1)` : append one (1) as its initial value, because all Pascal row starts with one
  - `for i in range(len(row)-1)` : we will iterate according to the length of `row`. `row` is an argument for the `count_pascal()` function.
  - `counting = row[i] + row[i+1]` : to count the middle Pascal value (between first 1 and last 1) by addition between current index (i) and the next index (i+1)
  - `new_list.append(counting)` : append the `counting` value to `new_list`
  - `new_list.append(1)` : add this just after the `for` loop but outside of it, this is a closing one (1), like was stated above, Pascal row always start with 1 and end with 1
  - `return new_list` : return the `new_list`


```python
pascal_depth = 5
def find_pascal(depth):
    default_list = [1]
    result = []
    for _ in range(depth):
        result.append(default_list)
        default_list = count_pascal(default_list)
    return result

def count_pascal(row):
    new_list = []
    new_list.append(1)
    for i in range(len(row)-1):
        counting = row[i] + row[i+1]
        new_list.append(counting)
    new_list.append(1)
    return new_list
print(find_pascal(pascal_depth))
```
