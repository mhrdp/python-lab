```python
# Recursion to count a sum of a list of number

"""
The way this work is,
1. Compute the length of the list and see if it was equal to one
  a. If True --> return the value, represent by n[0]
  b. If False --> return the first value (n[0]) + calling the function again (recursion)
    i. When calling the function again, the list inside the argument will start from the second value
    ii. This was representated by: sum_list_recursion(n[1:])
    iii. The new function calling will return yet another first value of the new list inside the new function calling's parameter.
2. So by the end of every function calling, it essentially return:
  n[0] + n[0] ... + n[0]
3. But, the last function calling actually outside of the addition function, right? Because it was in 'if' statement, not 'else'
  a. You would think that way, but remember that the last function call was called WITHIN the return inside of 'else' statement
  b. So it structured like this: n[0] + sum_list_recursion(n[1:]) + sum_list_recursion(n[1:]) ... + sum_list_recursion(n[1:]) <-- last function calling which will return n[0] because the there's only one item inside the list
"""

def sum_list_recursion(n):
  if len(n) == 1:
    return n[0]
  else:
    return n[0] + sum_list_recursion(n[1:])

sum_list_recursion([1, 2, 3, 4])
```

```python
# Recursion list sum

"""
1. This function will first iterate through the items inside the list 1-by-1
2. Then it will look for the type of each item,
  a. If the type is list, it will do the recursion and add it to the total variable value
    i. When the recursion was triggered, the 'for' method will 'break down' the list and iterate through it 1-by-1
    ii. Then it will return the sum of all item inside the list and add it to total variable
    iii. After it done iterate through all of the items, it will back to outside of the list and check for the item's type again
  b. If the type is any other than list, it will just do a normal addition
"""

def list_sum(n):
  total = 0
  for element in n:
    if type(element) == type([]):
      total = total + list_sum(element)
    else:
      total = total + element
  return total

list_sum([1, 2, [3, 4], [5, 6]])
```

```python
def factorial(n):
  if n <= 1:
    return 1
  else:
    return n * (factorial(n-1))
factorial(5)
```

```python
def additional(n):
  if n == 0:
    return 0
  else:
    return n + additional(n-2)
additional(10)
```

```python
# Explanation : 
# https://www.w3resource.com/python-exercises/data-structures-and-algorithms/python-recursion-exercise-6.php

def sum_of_num(n):
  if n == 0:
    return 0
  else:
    return n % 10 + (sum_of_num(n // 10))
sum_of_num(45)
```

```python
def geometric_series(n):
  if n < 2:
    return 1
  else:
    return 1/n + (geometric_series(n-1))
geometric_series(7)
```

```python
def harmonic_series(n):
  if n < 0:
    return 0
  else:
    return 1/(pow(2, n)) + harmonic_series(n-1)
harmonic_series(7)
```

```python
def count_power_of_with_recursion(n, p):
  if p == 0:
    return 1
  else:
    return n * (count_power_of_with_recursion(n, p-1))
count_power_of_with_recursion(3, 3)

# Alternatively, you could use pow(x, y) or x**y
# Or you could use pow(x, y, z) or x**y % z
```

```python
def recurgcd(lo, hi):
  low = min(hi, lo)
  high = max(hi, lo)

  if low == 0:
    return high
  elif low == 1:
    return 1
  else:
    return recurgcd(low, high % low)
recurgcd(12, 14)
```
