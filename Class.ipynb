```python
class ClassOne:
    def __init__(self, val1, val2):
        self.val1 = val1
        self.val2 = val2
    
    def call_back(self):
        return self.val1

class ClassTwo(ClassOne):
    def __init__(self, val1, val2):
        super().__init__(val1, val2)
        self.val3 = 1000
    
    def call(self):
        return self.val3
```

```python
val2 = ClassTwo(100, 200)

print(f'Call a function from child class, result: {val2.call()}')
print(f'Call a function from parent\'s class, result: {val2.call_back()}')
```

```python
class Build:
    def __init__(self, length, width, height):
        self.width = width
        self.length = length
        self.height = height
    
    def rect_area(self):
        return self.width * self.length
    
    def count(self):
        return self.width + self.length + self.height

class Rectangle(Build):
    def __init__(self, length):
        # This super() take an argument of length from child class
        # So the structure become something like `super().__init__(length=length, width=length, height=1)`
        super().__init__(length, length, 1)
    
    def count(self):
        # self.width here is equal to length parameter of Rectangle class
        return self.length * self.width * 10

class RandomBuild(Rectangle):
    # Because this class inherited from Rectangle, 
    # any value that passed to this class will be processed by Rectangle class __init__
    def count_a(self):
        rand_count = super().rect_area()
        return rand_count * 5

    def count_something(self):
        # This super method calls for __init__(self, length) from Rectangle class
        # So you only need to supply it with one argument
        # This super class will call for count() in Build class, ignoring the count() inside Rectangle class
        # Doing something like this generally not recommended, because this could be an indicative of design problem
        something = super(Rectangle, self).count()
        return something + 10
```

```python
rect_a = Rectangle(5)
print(f'Return value of count() inside Rectangle subclass: {rect_a.count()}')

rect = Rectangle(5)
print(f'Return value of build.rect_area() from Rectangle subclass: {rect.rect_area()}')

rand_build = RandomBuild(5)
print(f'Return value of RandomBuild.count() extended from Rectangle class: {rand_build.count_a()}')

rand_build_2 = RandomBuild(5)
print(f'Return value of count_something() function that inherit count() function from Build class: {rand_build_2.count_something()}')
```
