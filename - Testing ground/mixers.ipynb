```python
nums = [5,7,7,8,8,10]
target = 8

start = 0
end = len(nums)-1

i = 0
while i < len(nums):
    if nums[start] != target:
        start += 1
    if nums[end] != target:
        end -= 1
    i += 1

print(start, end)
```
